```python
import whisper

print("Loading Whisper Model...")
model = whisper.load_model("base")

# UPDATED: We are now using .mp3
audio_filename = "test_audio.mp3" 

print(f"Transcribing {audio_filename}...")

# This checks if the file actually exists before running to avoid errors
import os
if os.path.exists(audio_filename):
    result = model.transcribe(audio_filename)
    print("--- SUCCESS! HERE IS WHAT THE AI HEARD: ---")
    print(result["text"])
else:
    print(f"ERROR: I cannot find '{audio_filename}'. Check the file name in the sidebar!")
```
